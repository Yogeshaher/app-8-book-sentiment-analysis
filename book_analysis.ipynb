# what are most used word

```python
pattern = re.compile("[a-zA-Z]+")
finding = re.findall(pattern, book.lower())
finding[:5]
```

```python
d = {}
for word in finding:
    if word in d.keys():
        d[word] = d[word] + 1
    else:
        d[word] = 1
```

```python
d_list = [(value, key) for (key, value) in d.items()]
sorted(d_list, reverse=True)
```

```python
d_list = [(value, key) for (key, value) in d.items()]
d_list = sorted(d_list, reverse=True)
d_list
```

# sentiment analysis
## what is the most positive and the most negative chapter?

```python
from nltk.sentiment import SentimentIntensityAnalyzer
```

```python
analyzer = SentimentIntensityAnalyzer()
```

```python
 scores = analyzer.polarity_scores(book)
```

```python
if scores["pos"] > scores["neg"]:
    print("It is positive text")
else:
    print("It is negative text")
```

# chapters sentiments analysis

```python
import re
pattern = re.compile("Chapter [0-9]+")
chapters =  re.split(pattern, book)
chapters = chapters[1:]
```

```python
for nr, chapter in enumerate(chapters):
    scores = analyzer.polarity_scores(chapter)
    print(nr+1, scores)
```
